```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import LabelEncoder
from sklearn.svm import LinearSVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
```

```python
data_train=pd.read_csv("/Users/hussamuddin/Downloads/Genre Classification Dataset/train_data.txt",delimiter=":::",header=None,names=['ID', 'TITLE', 'GENRE', 'DESCRIPTION'])
x_test=pd.read_csv("/Users/hussamuddin/Downloads/Genre Classification Dataset/test_data.txt",delimiter=":::",header=None,names=['ID', 'TITLE', 'DESCRIPTION'])
y_test=pd.read_csv("/Users/hussamuddin/Downloads/Genre Classification Dataset/test_data_solution.txt",delimiter=":::",header=None,names=['ID', 'TITLE', 'GENRE', 'DESCRIPTION'])
```

```python
data_train['DESCRIPTION'].fillna("", inplace=True)
x_test['DESCRIPTION'].fillna("", inplace=True)

t_v = TfidfVectorizer(stop_words='english', max_features=100000)
X_train = t_v.fit_transform(data_train['DESCRIPTION'])
X_test = t_v.transform(x_test['DESCRIPTION'])

label_encoder = LabelEncoder()
y_train = label_encoder.fit_transform(data_train['GENRE'])
y_test = label_encoder.transform(y_test['GENRE'])
```

```python
X_train_sub, X_val, y_train_sub, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=42)

clf = LinearSVC()
clf.fit(X_train_sub, y_train_sub)

y_val_pred = clf.predict(X_val)
print("Validation Accuracy:", accuracy_score(y_val, y_val_pred))
print("Validation Classification Report:\n", classification_report(y_val, y_val_pred))
```

```python
y_pred = clf.predict(X_test)
print("Test Accuracy:", accuracy_score(y_test, y_pred))
print("Test Classification Report:\n", classification_report(y_test, y_pred))
```
